part2: Visualize the datasets using metplotlib libraries

```python
#create the datasets
people =['ajay','varun','akash','arun','arjun']
age = [23, 24, 25, 18,28]
weight = [70,71,72,73,74]
height = [170,171,172,174,175]
```

```python
import matplotlib.pyplot as plt
```

```python
#find the relationship between weight to height
#generate a scatter plot to visualize the relationship
plt.scatter(weight, height)
plt.xlabel('weight')
plt.ylabel('height')
plt.title('weight vs height')
plt.show()
```

```python
#generate the bar chart for people to their weight
plt.figure(figsize=(5,4))
plt.bar(people, weight)
plt.xlabel('people')
plt.ylabel('weight')
plt.title('people vs weight')
plt.show()
```

```python
#generate the histrogram between people age to their height
plt.figure(figsize=(5,4))
plt.hist(weight, height)
plt.xlabel('weight')
plt.ylabel('height')
plt.title('weight vs height')
plt.show()
```
